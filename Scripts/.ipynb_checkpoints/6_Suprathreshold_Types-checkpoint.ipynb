In the previous notebook we saw different suprathreshold response across our calibrated cells, which showed that specific cell types can selectively entrain on prefered subthreshold frequencies (ES). Here we examine if this can be explained by the cells' state (firing rate) alone, or if it is due to unique ion parameters that got optimized for each type.

For each cell, we gather their calibrated rate and the maximum presented MMR (across field rotations) at the maximum ES amplitude, for each ES frequency. This way we can plot MMR vs cell state per ES frequency.

```python
Plot_MMR_vs_FR ()
```

We see that cells with high FR are generally entrained less, and only at high frequencies. While slow FR cells prefer lower frequencies. This is true regardless of the cell type. This is more clear if we bin our cells in groups.

```python
Plot_MMR_vs_FR (grouped = True)
```

```python
Plot_MMR_vs_FR (grouped = True, annotate = True)
```

We see that slow FR cells prefer 8Hz, mid FR cells like all ES frequencies, and high FR cells prefer 140Hz. This is true regardless of the cell type. However, individual cell types present differences in their FR. Thus, we can group again per cell type.

```python
Plot_MMR_vs_FR (grouped = True, bytype = True, annotate = False)
```

We see that excitatory cells are generaly slower (FR) and prefer 8Hz, with only ITL23 giving some more entrainment at 28Hz. Inhibitory cells prefer 140Hz, since they are fast spikers, but we can see that SST are more entrained already at 8Hz, while PVALB and VIP respond mostly at 28 and 140Hz.

At this point, we are done with all subthreshold, passive, active, amplitude, phase, field shape, field amplitude, field frequency, field distance, suprathreshold entrainment. And we have found that passive cable theory can predict all subthreshold effects, but if cell entrains on the subthreshold effects is strongly related to the frequency relation between its firing rate and the ES frequency. However if the two frequencies are close, passive cable theory can predict the level of entrainment. Since the FR of individual cell types is more or less within a specified region, this can provide cell type targeted effects through ES frequency.
